```python
import PyPDF2
import pandas as pd

# Open the pdf file
pdf1 = PyPDF2.PdfFileReader('Resources/pubmed.pdf')
pdf2 = PyPDF2.PdfFileReader('Resources/pubmed2.pdf')
pdfs = [pdf1, pdf2]
# Create a dataframe to store the data from each page
texts=[]
names=[]
count=1
for pdf in pdfs:
    name = f'pdf{count}'
    count+=1
    # Get the number of pages in the pdf file
    num_pages = pdf.getNumPages()

    # Iterate over the pages in the pdf file
    for i in range(num_pages):

        # Get the text from the current page
        text = pdf.getPage(i).extractText()

        # Split the text into sections
        sections = text.split('\n\n')
        texts.append(sections[0])
        names.append(name)

# Add the section to the dataframe
pubmed_df = pd.DataFrame({'doc': names, 'page_text': texts})
pubmed_df
```
